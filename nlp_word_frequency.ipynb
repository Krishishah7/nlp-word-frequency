# NLP Word Frequency Analyzer

This project demonstrates word frequency analysis in Natural Language Processing (NLP) using Python.


```python
import nltk
import string
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from collections import Counter

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('punkt_tab')
```

```python
text = """
Natural Language Processing helps computers understand human language.
Word frequency analysis is a basic NLP technique.
It helps identify the most common words in a text.
"""

print("Original Text:")
print(text)
```

```python
# Lowercase
text = text.lower()

# Remove punctuation
text = text.translate(str.maketrans('', '', string.punctuation))

# Tokenization
tokens = word_tokenize(text)

# Remove stopwords
stop_words = set(stopwords.words('english'))
filtered_tokens = [word for word in tokens if word not in stop_words]

print("Filtered Tokens:")
print(filtered_tokens)
```

```python
word_freq = Counter(filtered_tokens)

print("Word Frequency:")
for word, count in word_freq.items():
    print(f"{word} : {count}")
```

```python
print("\nTop 5 Most Frequent Words:")
print(word_freq.most_common(5))
```
